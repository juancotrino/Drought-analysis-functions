```python
import numpy as np
from datetime import datetime, timedelta
from osgeo import gdal, gdal_array
```

```python
start = datetime.strptime("01-01-1979", "%d-%m-%Y")
end = datetime.strptime("31-12-1981", "%d-%m-%Y")
analysis_period_years = end.year - start.year
date_generated = [start + timedelta(days=x) for x in range(0, (end-start).days + 1)]
```

```python
src_filepath = "/Users/juanmanuel/Documents/Juan Manuel/Universidad/TESIS/Datos/meteodata/calib/prec/TIFF/"
dst_filepath = "/Users/juanmanuel/Documents/Juan Manuel/Universidad/TESIS/Datos/meteodata/calib/prec/TEST/"
shpfile_clip = '/Users/juanmanuel/Documents/OneDrive - ESCUELA COLOMBIANA DE INGENIERIA JULIO GARAVITO/1. TRABAJO DE GRADO/1. Tesis/2. Datos/0. Datos 20190912/Datos _iniciales/data/GIS/Mapa_General/cuenca_magdalena_proyectado.shp'
```

```python
reference_raster = gdal.Open("/Users/juanmanuel/Documents/Juan Manuel/Universidad/TESIS/Datos/meteodata/calib/prec/TIFF/daily-prec_19790101.tiff")
reference_raster_array = np.array(reference_raster.GetRasterBand(1).ReadAsArray())
x_pixels = len(reference_raster_array[0])  # number of pixels in x
y_pixels = len(reference_raster_array)  # number of pixels in y
```

```python
drought_binary_db = np.load("/Users/juanmanuel/Documents/Juan Manuel/Universidad/TESIS/Datos/meteodata/calib/prec/drought_binary_db.npy")
```

```python
year = 0
day_cnt = 0
driver = gdal.GetDriverByName('GTiff')
# follow code is adding GeoTranform and Projection
geotrans = reference_raster.GetGeoTransform()  #get GeoTranform from existed 'data0'
proj = reference_raster.GetProjection() #you can get from a exsited tif or import 
gdal.GDT_
for date in date_generated:
    fulldate = int(date.strftime("%Y%m%d"))
    dst_filename = dst_filepath + "daily-binary_drought_" + str(fulldate) + ".tiff"
    dataset = driver.Create(dst_filename,x_pixels, y_pixels, 1,gdal.GDT_Float32)
    dataset.GetRasterBand(1).WriteArray(drought_binary_db[:, :, year, day_cnt])
    dataset.SetGeoTransform(geotrans)
    dataset.SetProjection(proj)
    gdal_array.SaveArray(drought_binary_db[:, :, year, day_cnt], dst_filename, format = "GTIFF", prototype = reference_raster)    
    gdal.Warp(dst_filename,
              dataset,
              srcSRS='EPSG:21897',
              dstSRS='EPSG:21897',
              cutlineDSName=shpfile_clip,
              cropToCutline=True,
              dstNodata=-1)
    dataset.FlushCache()
    dataset = None
    if date.year != (date + timedelta(days = 1)).year:
        year += 1
        day_cnt = 0
        continue
    if date.day == 28 and (date + timedelta(days = 1)).day == 1 and date.month == 2:
        day_cnt += 2
        continue
    day_cnt += 1
reference_raster = None
```
